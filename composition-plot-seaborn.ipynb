```python
import pandas as pd
import seaborn as sns
from ordered_set import OrderedSet

sns.set()
sns.set_style('whitegrid')
sns.set_color_codes()
```

```python
tran = pd.read_csv("table_even20000_L2.txt", sep='\t', skiprows=1, index_col=0)
tran.head()
```

```python
description = list()
for i in tran.index.values:
    if i.startswith("G"):
        description.append("Gil")
    else:
        description.append("American")
tran["description"] = description
```

```python
groups = tran.groupby(tran.description).sum()
groups
```

```python
grp = groups.apply(lambda x: x / x.sum() * 100, axis=1)
grp
```

```python
Gil = grp.loc["Gil", :].nlargest(5)
American = grp.loc["American", :].nlargest(5)
```

```python
Gil
```

```python
American
```

```python
keys = list(OrderedSet(list(Gil.index) + list(American.index)))
keys
```

```python
final = grp[keys]
```

```python
final.plot.bar(position=1).legend(bbox_to_anchor=(1, 1))
```

```python
f_t = final.T
f_t.plot.bar(position=0).legend(bbox_to_anchor=(1, 1))
```

```python
f_t.to_csv("top5.phylum.csv")
```
